# Password generator

### Goal:
- Generate a random, secure password using letters, numbers and symbols
- Ask user for password length
- Build the password randomly

```python
import random
import string


length = int(input("How long should the password be?"))

l_letters = string.ascii_lowercase
u_letters = string.ascii_uppercase
numbers = string.digits
symbols = string.punctuation

char_pool = l_letters + u_letters + numbers + symbols

password_l = []

for i in range(length):
    password_l.append(random.choice(char_pool))

password = "".join(password_l)
print('\nYour generated password is:',password)
```
